```python
import pandas as pd
```

```python
dataset = pd.read_csv('dataset/Recon OS Scan.csv')
```

```python
dataset
```

```python
dataset2 = pd.read_csv('dataset/Recon Ping Sweep.csv')
```

```python
dataset2
```

```python
dataset3 = pd.read_csv('dataset/Recon Vulnerability Scan.csv')
```

```python
dataset3
```

```python
dataGabungan = pd.concat([dataset, dataset2, dataset3], ignore_index=False)
```

```python
dataGabungan
```

```python
x = dataGabungan.iloc[:, 7:76]
```

```python
y = dataGabungan.iloc[:, 83:84]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=30)
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
modelDtree = DecisionTreeClassifier(criterion='entropy', splitter='random')
```

```python
modelDtree.fit(x_train, y_train)
```

```python
y_pred = modelDtree.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
```

```python
accuracy
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.Figure(figsize=(12, 8))
```

```python
tree.plot_tree(modelDtree, feature_names = x.columns.values, class_names = np.array(['Recon OS Scan','Recon Ping Sweep','Recon Vulnerability Scan']), filled = True)
```

```python
import seaborn as sns
from sklearn import metrics
```

```python
target = np.array(['Recon OS Scan','Recon Ping Sweep','Recon Vulnerability Scan'])
```

```python
cMatrix = metrics.confusion_matrix(y_test, y_pred)
```

```python
plt.Figure(figsize=(15,15))
```

```python
sns.heatmap(cMatrix, annot=True, xticklabels=target, yticklabels=target)
plt.xlabel('Prediksi')
plt.ylabel('Fakta')
plt.show()
```
